```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split, KFold
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import os
```

```python
# Define constants
BATCH_SIZE = 16
IMAGE_SIZE = (256, 256)
EPOCHS = 100
NUM_CLASSES = 87  # Adjust according to your dataset
```

```python
# Directories for training and testing data
train_data_dir = 'C:/Users/lasit/Desktop/Campus/4th year/1sem/DL/dl_assignment/gemstones-images/train'
test_data_dir = 'C:/Users/lasit/Desktop/Campus/4th year/1sem/DL/dl_assignment/gemstones-images/test'
```

```python
# Data Augmentation
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip('horizontal_and_vertical'),
    tf.keras.layers.RandomRotation(0.2),
    tf.keras.layers.RandomZoom(0.2),
    tf.keras.layers.RandomContrast(0.2),
    tf.keras.layers.RandomBrightness(0.2),
    tf.keras.layers.GaussianNoise(0.1),
    tf.keras.layers.RandomTranslation(0.1, 0.1),
])
```

```python
# Load dataset paths and labels
file_paths = []
labels = []
```

```python

# Create a mapping of classes to their corresponding labels
class_names = os.listdir(train_data_dir)
class_indices = {class_name: idx for idx, class_name in enumerate(class_names)}
```

```python
# Populate the file_paths and labels lists
for class_name in class_names:
    class_dir = os.path.join(train_data_dir, class_name)
    for file in os.listdir(class_dir):
        if file.endswith(('.png', '.jpg', '.jpeg')):  # Adjust according to your image formats
            file_paths.append(os.path.join(class_dir, file))
            labels.append(class_indices[class_name])

# Convert to numpy arrays
file_paths = np.array(file_paths)
labels = np.array(labels)
```

```python
# Calculate class weights
class_weights = compute_class_weight('balanced', classes=np.unique(labels), y=labels)
class_weights_dict = {i: class_weights[i] for i in range(len(class_weights))}
```

```python
# Cross-validation setup
kf = KFold(n_splits=5, shuffle=True, random_state=42)
fold_accuracies = []
```

```python

# Cross-validation
for train_index, val_index in kf.split(file_paths):
    train_fold_files, val_fold_files = file_paths[train_index], file_paths[val_index]
    train_fold_labels, val_fold_labels = labels[train_index], labels[val_index]

    # Create TensorFlow Datasets for the current fold
    train_data = tf.data.Dataset.from_tensor_slices((train_fold_files, train_fold_labels))
    val_data = tf.data.Dataset.from_tensor_slices((val_fold_files, val_fold_labels))

    # Create a function to load and preprocess images
    def load_and_preprocess_image(path, label):
        image = tf.io.read_file(path)
        image = tf.image.decode_jpeg(image, channels=3)
        image = tf.image.resize(image, IMAGE_SIZE)
        image = data_augmentation(image)  # Apply data augmentation
        image /= 255.0  # Normalize to [0, 1]
        return image, label

    # Map the load and preprocess function
    train_data = train_data.map(load_and_preprocess_image, num_parallel_calls=tf.data.AUTOTUNE).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
    val_data = val_data.map(load_and_preprocess_image, num_parallel_calls=tf.data.AUTOTUNE).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

    # Load base model (InceptionV3) with pre-trained weights
    base_model = tf.keras.applications.InceptionV3(weights='imagenet', include_top=False, pooling='avg', input_shape=(256, 256, 3))
    base_model.trainable = False  # Freeze the base model

    # Define the complete model
    model = tf.keras.Sequential([
        base_model,
        tf.keras.layers.Dense(256, activation='relu'),
        tf.keras.layers.Dropout(0.5),
        tf.keras.layers.Dense(NUM_CLASSES, activation='softmax')
    ])

    # Compile the model
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    # Callbacks
    early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
    model_checkpoint = tf.keras.callbacks.ModelCheckpoint('best_model.keras', save_best_only=True)  # Update file extension

    # Train the model on the current fold
    history = model.fit(
      train_data,
      epochs=EPOCHS,
      validation_data=val_data,
      class_weight=class_weights_dict,
      callbacks=[early_stopping, model_checkpoint]
    )


    # Evaluate the model on the validation set
    val_loss, val_accuracy = model.evaluate(val_data)
    fold_accuracies.append(val_accuracy)

# Print average accuracy across folds
print(f'Average validation accuracy across folds: {np.mean(fold_accuracies)}')
```

```python
# Load the best model
model.load_weights('best_model.keras')
```

```python
# Evaluate on the test set
test_data = tf.keras.utils.image_dataset_from_directory(
    test_data_dir,
    image_size=IMAGE_SIZE,
    batch_size=BATCH_SIZE
)
```

```python
# Normalize test data (no augmentation)
test_data = test_data.map(lambda x, y: (x / 255.0, y))
```

```python
test_loss, test_accuracy = model.evaluate(test_data)
print(f'Test accuracy: {test_accuracy}')
```

```python
# Predictions on test data
predictions = model.predict(test_data)
predicted_classes = np.argmax(predictions, axis=1)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

# Generate classification report with zero_division to handle undefined precision
test_labels = np.concatenate([y.numpy() for _, y in test_data])
print(classification_report(test_labels, predicted_classes, zero_division=0))  # Avoid warnings for undefined metrics

# Generate confusion matrix
print(confusion_matrix(test_labels, predicted_classes))
```

```python
# Visualizing results (optional)
def plot_history(history):
    plt.plot(history.history['accuracy'], label='train accuracy')
    plt.plot(history.history['val_accuracy'], label='validation accuracy')
    plt.title('Model accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend()
    plt.show()

plot_history(history)
```
